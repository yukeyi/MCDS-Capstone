```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
fixed = "fixed.nii"
fixed_image = sitk.ReadImage(fixed)
fixed_image_arr = sitk.GetArrayFromImage(fixed_image)
```

```python
fixed_slice = fixed_image_arr[:, 97,:]
plt.imshow(fixed_slice)
```

```python
import scipy.misc
scipy.misc.imsave('image.jpg', fixed_slice)
image = cv2.imread('image.jpg')
gray= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

sift = cv2.xfeatures2d.SIFT_create(5)
kp = sift.detect(gray,None)
for i in kp:
    print(i.pt)
img=cv2.drawKeypoints(gray,kp,image)
plt.imshow(img)
cv2.imwrite('sift_keypoints.jpg',img)
```
