```python
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
```

```python
def drawBox(img,height,width,left,top):

    im = np.array(Image.open(img), dtype=np.uint8)

    # Create figure and axes
    fig,ax = plt.subplots(1)

    # Display the image
    ax.imshow(im)

    # Create a Rectangle patch
    rect = patches.Rectangle((left,top),width,height,linewidth=1,edgecolor='r',facecolor='none')

    # Add the patch to the Axes
    ax.add_patch(rect)

    plt.show()
```

```python
def get_unit_data(unit_dir, sample=32, num_cols=4):
    image_names = sorted(os.listdir(os.path.join(unit_dir)))[:sample]
    entries = []
    for image_name in image_names:
        # remove the first 5 chars containing the image rank
        parts = image_name[5:].split('-')
        raw_name = '{}-{}.jpg'.format(parts[0], parts[1])
        raw_width, raw_height = Image.open(os.path.join('/pylon5/ac5616p/faqian/ddsm-visual-primitives/data/ddsm_raw/{}'.format(raw_name))).size
        box = parts[2].split('_')
        x = int(box[1][1:])
        y = int(box[0][1:])
        w = int(box[3][1:])
        h = int(box[2][1:])
        assert w == h
        height = 100 * float(h) / raw_height
        width = 100 * float(w) / raw_width
        left = 100 * float(x) / raw_width
        top = 100 * float(y) / raw_height
        style = 'position: absolute; height: {}%; width: {}%; left: {}%; top: {}%;'.format(height, width, left, top)
        entry = {
            'img_name': image_name,
            'style': style,
            'raw_name': 'raw/{}'.format(raw_name)
        }
        print(entry)
        drawBox('/pylon5/ac5616p/faqian/ddsm-visual-primitives/data/ddsm_raw/{}'.format(raw_name),h,w,x,y)
```
